## Import Packages

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from scipy.io import wavfile
import IPython.display as ipd
import os
```

```python
species_num = 10

os.chdir('/Users/amikano/Documents/MSDS/DS6050/Project/audio_files/' + str(species_num) + '/')

file_name = 'Epetersi2_Daniel1.WAV'

### Load data ###
y, sr = librosa.load(file_name)
onset_frames = librosa.onset.onset_detect(y=y, sr=sr, wait=1, pre_avg=1, post_avg=1, pre_max=1, post_max=1)
onset_times = librosa.frames_to_time(onset_frames)

plt.figure(figsize=(20, 5))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveplot', fontdict=dict(size=18))
plt.vlines(onset_times, 0,ymax=0.6, color='r', alpha=0.9, linestyle='--', 
label='Onsets') #
plt.xlabel('Time', fontdict=dict(size=15))
plt.ylabel('Amplitude', fontdict=dict(size=15))
plt.show()
```

## Cut audio and parse into frog calls

```python
cutoff_by_seconds = 100


os.chdir('/Users/amikano/Documents/MSDS/DS6050/Project/audio_files/' + str(species_num) + '/')
rate, data = wavfile.read(file_name)

os.chdir('/Users/amikano/Documents/MSDS/DS6050/Project/audio_files/' + str(species_num) + '_new/')
os.mkdir(file_name[:-4])

os.chdir('/Users/amikano/Documents/MSDS/DS6050/Project/audio_files/' + str(species_num) + '_new/' + file_name[:-4])
os.mkdir("audio_outputs")
os.chdir('/Users/amikano/Documents/MSDS/DS6050/Project/audio_files/' + str(species_num) + '_new/' + file_name[:-4] + "/audio_outputs")

counter = 1
for time in onset_times[onset_times < cutoff_by_seconds]:
    # the timestamp to split at (in seconds)
    split_at_timestamp_left = time-0.5
    split_at_timestamp_right = time+0.5

    # get the frame to split at
    split_at_frame_left = int(rate * split_at_timestamp_left)
    split_at_frame_right = int(rate * split_at_timestamp_right)

    # split
    split_wav = data[split_at_frame_left:split_at_frame_right] # split
    
    # filename
    filename_split = file_name + '_' +str(counter) + '.wav'

    # save the result
    wavfile.write(filename_split, rate, split_wav)
    counter += 1
```

## Save a spectrogram for each of the segmented audio files 

```python
dr = "/Users/amikano/Documents/MSDS/DS6050/Project/audio_files/" + str(species_num) + "_new/"

folders = os.listdir("/Users/amikano/Documents/MSDS/DS6050/Project/audio_files/" + str(species_num) + "_new")
folders.remove('.DS_Store')

for folder in folders:

    wavs = os.listdir(dr + folder + "/audio_outputs")

    for wav in wavs:
        # Load data
        y, sr = librosa.load(dr + folder + "/audio_outputs" + "/" + wav)

        # Create plot
        fig = plt.figure();
        plt.axis('off');
        S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        S_dB = librosa.power_to_db(S, ref=np.max)
        librosa.display.specshow(S_dB, x_axis='s', y_axis='mel', sr=sr)

        path3 = dr + folder + "/image_outputs/"
        if not os.path.exists(path3):
            os.makedirs(path3)

        # Save plot
        fig.savefig(path3 + wav [:-4]+ '.png');
        plt.close()
```
